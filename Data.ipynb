```python
import pandas as pd
```

```python

Boston_csv1 = "Resources/Boston_Listings.csv"
Boston_Listings = pd.read_csv(Boston_csv1)
Boston_csv2 = "Resources/Boston_Reviews.csv"
Boston_Reviews = pd.read_csv(Boston_csv2)
Seattle_csv1 = "Resources/Seattle_Listings.csv"
Seattle_Listings = pd.read_csv(Seattle_csv1)
Seattle_csv2 = "Resources/Seattle_Reviews.csv"
Seattle_Reviews = pd.read_csv(Seattle_csv2)
```

```python
Boston_merge = pd.merge(Boston_Listings, Boston_Reviews, on = "id")
Boston_merge
```

```python
Seattle_merge = pd.merge(Seattle_Listings, Seattle_Reviews, on = "id")
Seattle_merge
```

```python
Boston_merge_clean = Boston_merge[["id","name","host_id","access","house_rules","host_name","host_since","host_response_rate","thumbnail_url", "street","neighbourhood","property_type","room_type", "city", "state","price","accommodates","bathrooms","bedrooms","beds","square_feet","security_deposit","cleaning_fee","minimum_nights","first_review","last_review","review_scores_rating","review_scores_accuracy","review_scores_cleanliness","review_scores_checkin","review_scores_communication","review_scores_location","review_scores_value","cancellation_policy","reviews_per_month","reviewer_id"]]
Boston_merge_clean
```

```python
Seattle_merge_clean = Seattle_merge[["id","name","host_id","host_name","host_since","host_response_rate","thumbnail_url", "street","neighbourhood","property_type","room_type", "city", "state","price","accommodates","bathrooms","bedrooms","beds","square_feet","security_deposit","cleaning_fee","minimum_nights","first_review","last_review","review_scores_rating","review_scores_accuracy","review_scores_cleanliness","review_scores_checkin","review_scores_communication","review_scores_location","review_scores_value","cancellation_policy","reviews_per_month","reviewer_id"]]
Seattle_merge_clean
```
